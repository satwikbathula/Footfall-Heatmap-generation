```python
import cv2
import cvzone
import time
import math
import numpy as np
import torch  

cap = cv2.VideoCapture(r"C:\Users\bathu\people_-_6387 (540p).mp4")

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

model_path = "C:/Users/bathu/facial_recognition/yolov8l.pt"
model = YOLO(model_path)  
classNames = ['person']  
heatmap = np.zeros((frame_height, frame_width), dtype=np.float32)

out = cv2.VideoWriter('cumulative_heatmap.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 20.0, (frame_width, frame_height))

fixed_max_value = 100  

while True:
    success, img = cap.read()
    if not success:
        break

    person_count = 0  # Counter for the number of people
    results = model(img, stream=True)

    for r in results:
        boxes = r.boxes
        for box in boxes:
            if box.conf[0] < 0.2:
                continue

            # Extract integer value from tensor
            class_id = int(box.cls.item()) if isinstance(box.cls, torch.Tensor) else int(box.cls)

            # Print class_id for debugging
            print(f"Class ID: {class_id}")

            if class_id < len(classNames) and classNames[class_id] == 'person':  # Check if the class label is 'person'
                person_count += 1  # Increment count for each person detected

                x1, y1, x2, y2 = box.xyxy[0]
                x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

                # Update the heatmap for 'person' only
                heatmap[y1:y2, x1:x2] += 1

                # Draw bounding box around 'person' only
                cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green color for bounding box

    # Display the person count on the image
    cv2.putText(img, f'People Count: {person_count}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

    heatmap_normalized = np.uint8(255 * np.minimum(heatmap, fixed_max_value) / fixed_max_value)
    heatmap_color = cv2.applyColorMap(heatmap_normalized, cv2.COLORMAP_JET)

    # Combine heatmap with the original image that now includes bounding boxes and person count
    overlayed_image = cv2.addWeighted(img, 0.8, heatmap_color, 0.2, 0)

    out.write(overlayed_image)

    cv2.imshow('Cumulative_test4 Heatmap Video', overlayed_image)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
